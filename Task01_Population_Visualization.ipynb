```python
male_2024 = male_df[["Country Name", "2024"]].dropna()
male_2024 = male_2024.sort_values(by="2024", ascending=False)
male_2024.head(10)
```

```python

female_2024 = female_df[["Country Name","2024 [YR2024]"]].dropna()
female_2024 = female_2024.sort_values(by="2024 [YR2024]", ascending=False)
female_2024.head(10)
```

```python
female_2024 = female_df[["Country Name", "2024 [YR2024]"]].copy()
female_2024.rename(columns={"2024 [YR2024]": "2024"}, inplace=True)
```

```python
female_2024
```

```python
female_df["2024 [YR2024]"] = pd.to_numeric(female_df["2024 [YR2024]"], errors="coerce")
```

```python
female_2024 = female_2024.sort_values(by="2024", ascending=False)
female_2024.head(10)
```

```python
total_population_df=total_population_df.rename(columns={"2024 [YR2024]": "2024"})
total_population_df.head(5)
```

```python
total_population_2024 =total_population_df[["Country Name","2024"]].dropna()
```

```python
total_population_2024["2024"] = pd.to_numeric(total_population_2024["2024"], errors="coerce")
```

```python
total_population_2024 = total_population_2024.sort_values(by="2024", ascending=False)
total_population_2024.head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=male_2024.head(10), x="2024", y="Country Name" ,palette="Set1")
plt.title("Top 10 Countries by Male Population Percentage (2024)")
plt.xlabel("Male %")
plt.ylabel("Country")
plt.show()
```

```python
female_2024["2024"] = pd.to_numeric(female_2024["2024"], errors="coerce")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=female_2024.head(10), y="2024", x="Country Name", palette="Purples_d")
plt.title("Top 10 Countries by Female Population Percentage (2024)")
plt.xlabel("Country")
plt.ylabel("Female %")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=total_population_2024.head(10), x="2024", y="Country Name", palette="Reds_d")
plt.title("Top 10 Countries by Total Population (2024)")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

# ✅ Final Insights for Population Visualization

## Insights – Male Population Percentage (2024)
The visualization shows the top 10 countries with the highest male population percentage in 2024.
Qatar, UAE, and Oman lead the chart, with male populations exceeding 60%, likely due to large-scale male migrant workforces.
## Insights – Female Population Percentage (2024)
The bar chart represents the top 10 countries with the highest female population percentage in 2024.
Hong Kong SAR, Moldova, and Macao SAR have over 53% female population, indicating a slightly female-dominant demographic in these regions.

## Insights – Total Population (2024)
The bar chart displays the top 10 economic and regional groupings by total population in 2024.
The World population has reached over 8.1 billion, with Low & Middle Income and Middle Income groups making up the majority share.
This emphasizes how much of the global population resides in developing and emerging regions.
